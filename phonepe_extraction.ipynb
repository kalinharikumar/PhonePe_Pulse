## ***Extraction of data from JSON to DataFrame***

```python
#extraction of aggregated-transaction data
import pandas as pd
import json
import os
aggregated_trans='/content/pulse/data/aggregated/transaction/country/india'
agg_trans_part=os.listdir(aggregated_trans)
df1=pd.DataFrame()
df2=pd.DataFrame()
for i in agg_trans_part:
  if i.isnumeric()==True:
    cur_year=aggregated_trans + '/' + i + '/'
    files=os.listdir(cur_year)
    for file in files:
      cur_file=cur_year + '/' + file
      file_path=open(cur_file,'r')
      quaters=json.load(file_path)
      for data in quaters['data']['transactionData']:
        name = data['name']
        count = data['paymentInstruments'][0]['count']
        amount = data['paymentInstruments'][0]['amount']
        df = pd.DataFrame([{'year':i,'quarter':int(file.strip('.json')),'paymentinstrument':name,'count':count,'amount':amount}])
        df1 = pd.concat([df1,df],axis=0)

  if i.isnumeric()!=False:
    state=aggregated_trans + '/' + 'state' + '/'
    state_list=os.listdir(state)
    for j in state_list:
      cur_state=state + '/' + j + '/'
      year_list=os.listdir(cur_state)
      for k in year_list:
        cur_year=cur_state + '/' + k + '/'
        files=os.listdir(cur_year)
        for file in files:
          cur_file=cur_year + '/' + file
          file_path=open(cur_file,'r')
          quaters=json.load(file_path)
          for data in quaters['data']['transactionData']:
             name = data['name']
             count = data['paymentInstruments'][0]['count']
             amount = data['paymentInstruments'][0]['amount']
             df = pd.DataFrame([{'state':j,'year':k,'quarter':int(file.strip('.json')),'paymentinstrument':name,'count':count,'amount':amount}])
             df2 = pd.concat([df2,df],axis=0)


df1_sorted=df1.sort_values(['year','quarter'], axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)
df2_sorted=df2.sort_values(['year','quarter'], axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)  

yearly_trans=df1_sorted.reset_index(level=None, drop=True, inplace=False, col_level=0, col_fill='index')
yearly_state_trans=df2_sorted.reset_index(level=None, drop=True, inplace=False, col_level=0, col_fill='index') 
  
```
